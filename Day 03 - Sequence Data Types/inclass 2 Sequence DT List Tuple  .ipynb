## SEQUENCE DATA TYPES

Following methods on a sequence types: LIST and TUPLE
- `len(list)` - To find the number of elements 
- `max(list)` - gives the highest element/values 
- `min(list)` - gives the lowest element/value

- `index(value)` - will reveal the index of an element
- `count(value)` - count of a particular value

- packing/ unpacking - 

- membership test - use `in` or `not in` operation to check whether something is inside a list or tuple
 
Work on Lists -
- `append()` - Add another element to the end of the list - at the last index
- `clear()` - removes everything
- `insert(index,value)` - insert an element at an index, and move other elements ahead - target *index,value*
- `pop()` - used to get and remove last element, but you can target *index*
- `remove(value)` - will delete the first occurence of an element - target *element/value*

- `reverse()` -  invert the order
- `sort()` - sort the elements in a list

Work on tuples - 
- shallow immutability - saving a mutable list within a tuple
- packing / unpacking - here you don't even need the ( )

## 3. Tuple

- Tuples are immutable
- cannot modify them, once made

````python
num_tuple = (10,20,30,40)
print(num_tuple)

# access using index
num_tuple[3] 

# modify using index
num_tuple[0] = 33 # DOES NOT WORK
# BECAUSE TUPLE IS IMMUTABLE
````

````python
idrees = (0, 'idrees', 45.55, 'card')
idrees[2] = 65
````

````python
# co ordinates of places

coords_KualaLumpur = (03.0852, 101.4143)
coords_KualaLumpur
````

#### - tuple packing / unpacking

````python
# bagging a couple of elements s into item variable
item = 11 , [1,2,"a"]

# list unpacking
a , b = item 

x , y, z = 100, 200, 300
````

````python
t = (1,2,3,4)

a,b,c,d = t

p,q,r,s = t
````

````python
# packing unpacking

x , y = 20 , 40
print(f"x: {x}, y: {y}")

y, x = x, y

print(f"x: {x}, y: {y}")
````
