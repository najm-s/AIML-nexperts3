## 3. Set

A set is simply a collection of distinct objects.
- un ordered - no first or last
- { } - in maths, we use the same notation for sets
- sets do not allow duplicates - auto removes duplicates
- unindexed - no indexes and cannot use [0] or [1] etc


- https://www.geeksforgeeks.org/maths/sets/
- [SET OPERATIONS](https://www.geeksforgeeks.org/maths/set-operations/)

```python
items = { 1,2,3,4,5}
```

- can bag multiple types in SETs

```python
itemsx = { 1,'2',3,'4'}
itemsx = { 1,'2',3,'4', (5,2) } # tuples
```

```python
itemsx = { 1,'2',3,'4', [5,2] } # list will need indexes
itemsx = { 1,'2',3,'4', (5,2), {'a':1,'b':3} } # dictionaries
itemsx = { 1,'2',3,'4', (5,2), {'a','b',3} } # sets
```

```python
items = {}
items
```

```python
list1 = [1,2,3,4,4,4,4,4,4,4,4,4,4]

# remove duplicates
set1 = set(list1)
list1 = list(set1)

# remove duplicates
list1 = list(set(list1))

list1
```

#### SET OPERATIONS

```python
a = {1,2,3,4}
b = {3,4,5,6}
```

```python
c = list(b)
c.reverse()
c.sort()

c
```

```python
# UNION
union_op = a | b
union_op = a.union(b)
union_op

# INTERSECTION
intersection_op = a & b
intersection_op = a.intersection(b)
intersection_op

# Difference
difference_op = a - b
difference_op = a.difference(b)
difference_op
```
