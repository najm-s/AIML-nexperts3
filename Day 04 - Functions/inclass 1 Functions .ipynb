## Functions

- functions - in mathematics - are simply things that take an input and give an output
- In programming:
    - Functions are reusable blocks of code that perform a specific task
        - organize our code
        - code readable
        - `reuse code` - don't need to write huge chunks of code for tasks
    - define a function:
        - define what `inputs` will be taken
        - what will happen (``process``) to the inputs
        - what will function `return`

#### Things related to functions

##### DEFAULT PARAMETERS

````python
# defualt paramters
def idrees(x=1,y=1,z=1):
    return x*y*z

idrees(2,2)
# we passed no arguments
# but we had defualt arguments there
# so no error
````

##### KEYWORD ARGUMENTS

````python
# keyword arguments paramters
# order does not matter
def idrees(x=1,y=1,z=1):
    result = (x + y)/z
    return result

idrees(z=5, x=10, y=30) # keyword arguments

idrees(2,2,1)
````
