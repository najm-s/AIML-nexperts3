## Recursion

- function calls itself 
- either directly or indirectly 
- you break a big problem into smaller versions of the same problem


```python
def function(parameters):
    if (CONDITION):
        return answer

    else:
        return function(with modified parameters)

````python
# simple countdown function 

def countdown(n):
    # base condition - which we use to stop recursion
    if n == 0:
        return n  # return terminates a function
    
    # problem solving using recursion
    print(n)
    countdown(n-1) # recursion

countdown(5)
````

````python
# calculate factorial
# n 
# x = 1 * i - 1,2, 3,4,5, ... n

def fact(n):
    x = 1
    for i in range(1, n+1):
        # x = x * i
        x *= i
    return x

fact(5)
````

````python
# factorial program with recursion
def fact(n):
    #base case
    if n == 0:
        return 1
    return n * fact(n-1)

fact(5)
````

````python
# create a function that adds all the previous numbers
def addRec(n):
    #base case
    if n == 0:
        return 0
    return n + addRec(n-1)

# addition of first 10 numbers
addRec(10)
````
