```python
# build a class called Vehicle
# blue print for vehicle - 
class Vehicle:
   def __init__(self, type, brand):
       self.type = type # attribute
       self.brand = brand
   # method
   def drive(self):
       print(f"{self.type} of {self.brand} is moving") # method
```

```python
class Dog:
    def __init__(self, name, age, birthday, tricks):
        self.name = name
        self.age = age
        self.birthday = birthday
        self.tricks = tricks
    def bark(self):
        print(self.name," says Woof!")
    def trick(self):
        print(self.name, " can do a ", self.tricks)
       
dog1 = Dog ("Rolex", 3, "2020-10-05", "back-flip")
dog2 = Dog ("Ted", 5, "2020-10-05", "squat")
 
dog1.bark()
dog2.bark()
 
dog1.trick()
dog2.trick()
```
