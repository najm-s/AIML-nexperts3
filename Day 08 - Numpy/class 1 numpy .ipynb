```python
# N - Dimensions

# 1 Dimenstional Array
list_1D = [x for x in range(10)]

array_1D = np.array(list_1D)

print(array_1D, "\nnumber of dimension is: ",array_1D.ndim)

# access elements within a list , [ ]
list_1D[1]
array_1D[4]
```

```python
# 2 Dimenstional Array
list_2D = [list_1D, list_1D]

array_2D = np.array(list_2D)

print(array_2D,"\nnumber of dimension is: ", array_2D.ndim)
print(list_2D)

list_2D[1][5]
```

```python
# 3 Dimenstional Array
list_3D = [list_2D, list_2D, list_2D]

array_3D = np.array(list_3D)

print(array_3D,"\nnumber of dimension is: ", array_3D.ndim)
print(list_3D)

list_3D[2][1][9]
```

### Array Methods

```python
# use np.arange

arr = np.arange(10)
print(arr)


# np.array_split()
arr_parts = np.array_split(arr, 3)
print("======arr_parts=====")
for part in arr_parts:
    print(part)
```

```python
# Linspace  -- evenly spaced values
# start, stop, num
# if 10,50, 5 --> start from 10, ending will 50 ---> INCLUSIVE 
line = np.linspace(10,50,5)
print("linspace of line: ", line)
```

```python
# zeros and ones methods
# pass any shape, ---> rows x columns
print("--------------Zeros Matrix---------------")
a = np.zeros((3,2,3))
print(a)

print("--------------Ones Matrix---------------")
b = np.ones((3,3))
print(b)

print("--------------Identity Matrix------------")
identiy_matrix = np.eye(3) # rows = columns identity matrix
print(identiy_matrix)
```

#### Random
- generation of random numbers
- random number 

```python
# set a seed -- choose a starting point for the generator function
np.random.seed(1)       # helps us make our results, Reproducable

# random values --- from [0, 1)]
random_floats = np.random.rand(2,3,3) 
print(random_floats)


# start point, end point --- radnom number between them --- of shape
random_integers = np.random.randint(0,9)            # random number from 0 to 9
random_integers = np.random.randint(0,9,(10,))      # random array of numbers
random_integers = np.random.randint(0,9,(3,3))
print("--------------random_integers------------")
print(random_integers)
```

#### Slicing

```python
# slice lists
list_1D[1:5] # 
list_1D[1:5:2] 
list_1D[1::]
list_1D[:]  

# list_2D[0, 1] # does not work
list_2D[0:2][:5]
# list does not really support submatrix

```

```python
# slice arrays
array_1D[1:5] # works 
array_1D[1:5:2] # start: stop: step  # works

array_1D[1::]
array_1D[:]

array_2D[0,] # [ 1st array, 2nd array  ]# works here

array_2D
```

```python
# rows columns
array_2D[0:2][:5]

# if using comma, think about rows, columns
array_2D[1 , :5]


# SUBMATRIX --- rows, columns
array_2D[: , :5] # all rows , only 0 to 5 columns


array_3D[1:,0:1,:5] # 1d array, 2d array, 3d array
```
