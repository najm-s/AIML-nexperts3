### Exercise

- Create dictionary name `data` of 5 `Students`
- Store `Math`, `English`, `Science` marks

```python
data = {
    'Student': ,
    'Math': ,
    'English': ,
    'Science': ,
}
```

```python
data = {
    'Student': ['Idrees','Saufi','Faiz','Dila','Nadirah'],
    'Math': [99, 85, 95, 85, 98],
    'English': [45, 35, 25, 65, 59],
    'Science': [55, 65, 75, 85, 87],
}

data
```

```python
# pip install pandas

import pandas as pd


# create a dataframe ---- Tables

df = pd.DataFrame(data)
print(df)
# dataframe appears as cool looking table in jupyter notebook

# index ---- column1 column2 ...

type(df)    # pandas.DataFrame
```

```python
# indexing ----- get the contents from an index
# df[0] # does not work here

# but it is working with columns
df['Student']       # works  #  pandas.Series ---- of Student column
df.Student   # works  #  pandas.Series ---- of Student column
```

```python
# series object ------ indexed sequences

srs = pd.Series(data)
srs

# index   ------ Content

# Series[index]

# srs[0] # does not work
srs['Student'] # Does work --- Contents of student
data['Student'] # Same thing, but we are using dictionary in this case!!!
```

```python
Math = [99, 85, 95, 85, 98]

# A proper series object
Math_srs = pd.Series(Math) # pandas.Series
Math_srs

# index --- actually is working like a charm
type(Math_srs) # exact result for us
type(Math_srs[2])

# Series --- saves on every index numpy data
```

```python
# df ---->  series
df.English # you can access each of the columns
df.Math
df.Student
```

### Data types in dataframes in Pandas
- There are five main dtypes in pandas:

    - `object`: Text or mixed numeric values
    - `bool`: True or False values
    - `int64`: Integer values
    - `float64`: Floating point values
    - `datetime64`: Date and time values

```python
# official documentation of pandas
# Create a sample DataFrame
df2 = pd.DataFrame({
   'float': [1.0],
   'int': [1],
   'datetime': [pd.Timestamp('20180310')],
   'string': ['foo']
})

# Check data types of all columns
print(df2.dtypes)
```

```python
df3 = pd.DataFrame({'Weight': [45, 88, 56, 15, None],
                    'Weight': [45, 88, 56, 15, 25],
                   'Name': ['Sam', 'Andrea', 'Alex', 'James', 'Idrees'],
                   'Name2': ['Sam', 'Andrea', ['Alex'], True, 12],
                   'Age': [14, 25, 55, 8, 'accidentaly a string was put here']})

print(df3.dtypes)
```

##### Methods useful for EDA - Exploratory Data Analysis
- .sort_values()
- .sort_index()
- .value_counts()
- .unique()
- .duplicated()
- 

```python
df
df_sortedby_english = df.sort_values(by='English', ascending=False)
df_sortedby_english
```

```python
df_sortedby_english.sort_index(ascending=True)
```

```python
# sorting just a column object ---- series object

df.English.sort_values()

df['English'].sort_values()
df['English'].sort_index()
```

```python
df_birds = pd.DataFrame(
    {"num_legs": [2, 4, 4, 6], "num_wings": [2, 0, 0, 0]},
    index=["falcon", "dog", "cat", "ant"],
)

df_birds
```

```python
# The value_counts() method in pandas is used to count the frequency of unique values in a DataFrame.
df.value_counts() # returns count of each value that appears in a column

df['Math'].value_counts()
df.Math.value_counts()
```

```python
# .unique() will return the unique values within a column ----- a series object --- not a data frame
df.English.unique()
df['Math'].unique()
```

```python
# .duplicated() 
# run this on df ---- returns True for a duplicate row
# run this on series ---- returns True for a duplicate Value
df.duplicated()

df.English.duplicated()
df.Math.duplicated()
```

```python
# .drop_duplicated()

df.Math.duplicated()
unique_math = df.Math.drop_duplicates()
unique_math
```
