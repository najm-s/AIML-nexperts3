```python
# Created A MINI Calculator
number1 = int(input("Enter first number:"))
number2 = int(input("Enter second number:"))

sum = number1 + number2
mul = number1 * number2
sub = number1 - number2
div = number1 / number2
fdiv = number1//number2 # rounds off the number to the smaller value
mod = number1 % number2 # Remainder

print(f"For the numbers: {number1} and {number2} \nSum: {sum}\nSubtraction: {sub}\nMultiplication: {mul}\nDivision: {div}")
print(f"Floor Division:{fdiv}")
print(f"Modular Division:{mod}")
```
