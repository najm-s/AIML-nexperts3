#### BarCharts in matplotlib
- Bar chart will be used for a catagorical variable
- all teh arguments that go into the `plt.bar()`
- univariate, bivariate - multivariate
- 

```python
products = ["Apple", "Mango", "Orange", "Banana"]
sales = [10 , 25, 45, 30]

# to make things easy
x = ["Apple", "Mango", "Orange", "Banana"]
y = [10 , 25, 45, 30]

import matplotlib.pyplot as plt

# plt.bar(x, y )
plt.bar(products, sales)
```

```python
# we get a lot of arguments with these methods
# some commonly used ones
# use whichever you need in any case


plt.bar(
    x= products,        # horizontal varaible
    height=y,           # height or vertical magnitudes
    width= 0.5,         # width of the bar
    bottom = None,      # stacked bar charts
    align= 'center',    # center your viz
    color = "#078776",     # use to change colors, # can use hex
    edgecolor = "#104B4B", # border color
    linewidth = 2,           # border thickness
    tick_label = ['a','b','c','d'],  # names for each bar
    hatch = '////\\' # x /\\ + o O
)


patterns = ['/', 'o', '*', '\\']
plt.bar(products, sales, hatch =patterns)
```

```python
x = ['Group A', 'Group B', 'Group C']
y = [20, 35, 30]
hatches = ['---', '+++', 'xxx']
colors = ['#f0f0f0', "#7b1010", "#491C93"] # Light to dark gray

plt.bar(x, y, color=colors, edgecolor='black', hatch=hatches)
plt.show()
```

```python
# Stacker Bar charts
x = ["Apple", "Mango", "Orange", "Banana"]
y = [10 , 25, 45, 30]
y1 = [10 , 25, 45, 30]

plt.bar(x,y, label = 'Online', color = "#15E654")

# we can do bi-variate analysis
plt.bar(x, y1, label = 'y1', bottom = y , color = "#9D2191")
```

```python
# multivariate analysis
import numpy as np

x = ["Apple", "Mango", "Orange", "Banana"]
y = np.array([10 , 25, 45, 30])
y1 = np.array([10 , 25, 45, 30])
y2 = np.array([10 , 25, 45, 30])

# we can do multi-variate analysis
plt.bar(x, y, label = 'y', bottom = y1 + y2, color = "#9D2191")
```

#### Histogram
- Used for a variable that is continous, plot its frequency distribution
- To check how many times something is appearing

- ![DD](https://www.mathsisfun.com/data/images/bar-chart-vs-histogram.svg)

```python
x = np.random.normal(20, 100, 1000)

plt.hist(x, bins= 20,  color='skyblue', edgecolor = 'black', alpha = 0.5 )

plt.show()
```

#### Pie Chart
- Filled circle display **distribution** of values over it
- in ML

```python
y = np.array([10 , 25, 45, 30])

labels = ["Apple", "Mango", "Orange", "Banana"]

explode = [0.2,0,0.1,0]

colors = ["#3CFF00", "#2FC700", "#1F8500", "#114B00"]


plt.pie(y, 
        labels= labels,
        explode=explode,
        shadow= True,
        autopct= '%1.2f%%', # to print how many percent
        pctdistance= 0.7, # radial distance of pct
        colors= colors
        )

plt.show()

#     python--- width of whole number     # numer of decimals      floating point         print the literal %
#autopct= '%----  1 -----------------------   .    1    ---------------- f  -------------    %%'
```

#### Boxplot
- understanding distribution of data
- Box and whisker plot 
- It divides our data into quartiles - is 25 % - a quarter
- show you the median

```python
x = np.random.randint(20, 100, 1000)
x

plt.boxplot(
    x,
    patch_artist=True,  # Fill with color
    boxprops= dict(facecolor = 'skyblue'),
    notch=True,         # Notched box
    vert=True,          # Vertical boxes
    labels ='x',  # X-axis labels
    showmeans=True      # Show mean marker
)
plt.show()
```

#### Scatter plot
- plots the points for representing the distribution of data for two variables

```python
x = [1,2,3,4,5,6,7]
y = [21,100,35,25,22,15,23]
bubble_sizes = [100, 200, 300, 400, 500, 600, 700 ]
y2 = [22,55,33,43,26,32,28]

plt.scatter(x, y, 
            s = bubble_sizes, # size of each point in scatterplot
            c = y,   # color
            cmap= 'inferno', # color map 
            alpha=0.6, 
            marker= '^', # x, +, *, v , ^
            edgecolors='black', 
            linewidths=2)

plt.scatter(x, y2, s = bubble_sizes, c = y2, cmap='viridis', marker= 'o', alpha=0.5) # inferno plasma magma viridis
```

#### Subplots 
- plots that have multiple visualizations

```python
cities = ["Kota Bharu", "Kuala Lumpur", "Penang", "Johor Bahru", "Kuching"] # catagorical

country_number = [i for i in range(5)] # numerical
 
# y is temperature values (C)
np.random.seed(42)
day1 = [temperature for temperature in np.random.randint(15, 35,5)]

# ONE BY ONE method

plt.subplots(1,3, # specifies, or creates a window
             figsize = (6, 2) # width x height
             )

# rows, columns, active_index
plt.subplot(1, 3, 1) # 1 row, 2 columns, first plot
plt.plot(country_number, day1)

plt.subplot(1, 3, 2) # 1 row, 2 columns, second plot
plt.bar(cities,day1)
plt.xticks(rotation = 45)

plt.subplot(1, 3, 3) # 1 row, 2 columns, second plot
plt.plot(cities,day1)

plt.show()
```

```python
cities = ["Kota Bharu", "Kuala Lumpur", "Penang", "Johor Bahru", "Kuching"] # catagorical
country_number = [i for i in range(5)] # numerical
# y is temperature values (C)
np.random.seed(42)
day1 = [temperature for temperature in np.random.randint(15, 35,5)]

# ALL AT ONCE

fig, ax = plt.subplots(1,3, # specifies, or creates a window
             figsize = (6, 2) # width x height
             )

# fig 1
ax[0].plot(country_number, day1)
ax[0].set_title('country day')

# fig 2
ax[1].bar(cities,day1)
ax[1].set_title('city day')

# fig 3
ax[2].plot(cities,day1)
ax[2].set_title('cities day')


plt.tight_layout()

plt.show()
```

```python
# dashboards

# ALL AT ONCE

fig, ax = plt.subplots(2,2, # specifies, or creates a window
             figsize = (16, 9) # width x height
             )

# fig 1
ax[0,0].plot(country_number, day1)
ax[0,0].set_title('country day')

# fig 2
ax[0,1].bar(cities,day1)
ax[0,1].set_title('city day')

# fig 3
ax[1,1].plot(cities,day1)
ax[1,1].set_title('cities day')
```
