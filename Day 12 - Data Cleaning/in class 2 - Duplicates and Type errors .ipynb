### **Duplicates**
- **Exact Duplicates**: 
    - Every single column in a row A is identical to row B
    - Often because of a *glitch* e.g. the user clicked on *submit form* 3 times, checked out from cart 4 times because of impatience. 
    - Easy to flag these

- **Partial Duplicates**:
    - Only specific columns match in the rows.
    - A bit more dangerous, because these can look like unique records

```python
# check for duplicates

df.duplicated().sum() # the number of duplicate records
```

```python
df.info()
```
