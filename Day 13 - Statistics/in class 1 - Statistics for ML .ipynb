```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

data = {
    'age': [10,20,30,40,50],
    'salary': [10000,20000,30000,40000,50000]
}

df = pd.DataFrame(data)
df
```

### Measure of Central Tendency
- what lies at the centre of the data
- what is the main part of data
- 

```python
df.mean()

df.median()

df.mode().iloc[0]


```

### Measure of Dispersion (Spread)
- Range
    - taking the `max - min`
- Variance 
    - $\sigma^{2}$ - (data - mean )/ N 
    - Variance is a huge number
    - There is no standardization in data, variance cannot really help us understand how spread out data is
    - 
- Standard Deviation 
    - $\sigma$ - square root of variance
    - this defines the window size, and if std is larger, data is spread about a lot


```python
df.var()

df.std()
```

### Scaling  - A part of Feature Engineering(data pre-processing)
1. **Standardization** 
- z = {frac\{$data$ - $\mu} / \sigma$}

- data - mean / std. dev
- if we standardize data, then look at the values, we can tell the dispersion by looking at how zscores are spread
- larger zscores i.e closer to 3 or more, means greater spread

- it makes sure that mean of whole data becomes 0
- Standard Deviation becomes 1


2. **Normalization**
- data_new = data - data_min / data_max - data_min
- 
- you squish the data between :
    - 0 to 1
    - or -1 to 1

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

std_scaler = StandardScaler()
minmax_scaler = MinMaxScaler()


standardized_df = pd.DataFrame(std_scaler.fit_transform(df), columns= df.columns)
standardized_df

normalized_df = pd.DataFrame(minmax_scaler.fit_transform(df), columns= df.columns)
normalized_df
```
