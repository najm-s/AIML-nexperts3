### **The ML Lifecycle**
- as a series of steps that we follow to build a machine learning system:
1. **Problem Scoping**:
    - Define what problem we are solving e.g. predicting whether a customer will keep our service or leave it - Customer Churn
2. **Data Acquisition**:
    - Collect relevant data from various sources - website, database, API, spreadsheet, 
3. **Data Exploration** - **Exploratory Data Analysis**:
    - We examine the data - if necessary, we clean it - we analyse the data using visualizations
4. **Modeling**
    - Select the relevant model - specify its *relevant parameters* (switches that control the behaviour of a model)
5. **Training**
    - We *split the data* we have into two parts - training, other for testing - Also called the `Train-Test Split`
    - We fit the model with data - to train it
6. **Testing**
    - We calculate various performance metrics to test how well our model is doing ---- *model performance*
7. **Deployment**
    - Deploy the model somewhere people can use it e.g. create custom UI and deploy, AWS, Azure, GCP
- 
-------- if performance is great, we move onwards ------ otherwise we repeat from any previous step ------------------
- That is why the whole thing is called a LifeCYCLE
- resplit the data ---- tune the hyperparameters ---- EDA again and check if data is good ---- data cleaning ----- change the data ---- add more data

#### **4. Modeling**
- modeling in Ml means choose a model
- choose its best parameters
    - data modeling is a concept in database, analytics etc

```python
# select the model

# Decision Tree Model
# tree type models

from sklearn.tree import DecisionTreeClassifier 
# DT classifier is a classification type model
# our input can be divided into different classes

model = DecisionTreeClassifier(
    # model hyperparameters
    max_depth= 3,# max_depth ----> one of those switches that will define how our model will behave
    criterion= 'gini'  # 'gini' 'entropy' 'log_loss'
    )
```

#### **5. Training**
- 
- each flower was given a `label` of species by a botanist
- `species` columns =======> `target` feature
- `sepal length (cm)	sepal width (cm)	petal length (cm)	petal width (cm)` columns =======> `input` features
- 
- commonly accepted Train Test ratio : 8:2
    - 80% data =======> Training
    - 20% data =======> Testing

```python
# select a target feature
# feature selection =======> select the input features, and the output feature
X = irisdf.drop('species', axis = 1 ) # input features ==== sepal length (cm),	sepal width (cm),	petal length (cm),	petal width (cm)
# two dimensional ==== dataframe

y = irisdf['species'] # target features ===== species
# one dimensional ==== series

# Train-Test Split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, # X : [[]] ====== y: []
                                                    test_size= 0.2, # let's use 20% of data for Testing
                                                    random_state= 42, # this sets the seed to choose from random rows
)


# tuple unpacking????
```

```python
print("X-train: ", X_train.shape, "=============>   training")
print("X-test: ", X_test.shape, "=============>  feature used for Testing")

print("y-train: ", y_train.shape, "=============>  training")
print("y-test: ", y_test.shape, "=============>  feature used for Testing")
print("=================================================================")
print("Training: these many records/rows ", X_train.shape[0])
print("Testing: these many records/rows ", X_test.shape[0])
```

```python
# feeding data to the model
# training the model on labelled data

model.fit(X_train, y_train) # Training data into model ====== model will learn patterns from the y_train ----- labels
```

#### **6. Testing**
- let's test the performance of the model
- 
- we use some classification metrics, or measures of performance
- Accuracy ===== right/ all the predictions

```python
y_predictions = model.predict(X_test)


from sklearn.metrics import accuracy_score

accuracy_DT = accuracy_score(y_test, y_predictions) # real values ===== predicted values

print(f'Decision Tree Model accuracy: {accuracy_DT*100:.2f}%')

# print("Decision Tree Model accuracy:", accuracy_DT*100,"%")
```

Step 2: We deploy our model

```python
# define a function that makes this prediction or uses this model
def predict_specie(sepal_L, sepal_W,petal_L,petal_W):

    target_names = ['setosa','virginica','versicolor']

    input_data = [[sepal_L,sepal_W,petal_L,petal_W]]

    predictions = model.predict(input_data)
    # print(input_data)
    # print(target_names[predictions[0]])

    return target_names[predictions[0]]
```

### Revise the whole pipeline at once:
- Important Steps 

```python
# step 1: we will predict flower species, based on some physical measurements
# step 2: getting data
from sklearn.datasets import load_iris
iris = load_iris()

# step 3: data explore
import pandas as pd
irisdf = pd.DataFrame(iris.data, columns= iris.feature_names)

irisdf['species'] = iris.target

# checked number of values, checked for missing values
# irisdf.info()
irisdf.isna().sum()
# you can do visuals to check class balance as well

# step 4: modeling
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(random_state= 42) # keeping parameters as default # set the seed ===> helps reproduce results

#step 5: training
# feature selection
X = irisdf.drop('species', axis= 1) # setting input features
y = irisdf.species

# traintest split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state= 37) # set the seed # ====> helps reproduce outputs

# train the model === fit the model with data
model.fit(X_train, y_train)

# step 6: Testing
# make predictions on some labelled input
y_predictions = model.predict(X_test)

# calcualte accuracy
from sklearn.metrics import accuracy_score
print(f"Model accuracy is====>  {accuracy_score(y_test, y_predictions)*100:.2f}%")
```
