### 4.  String

- write alphabets, or sentences 

- use quotes, single or doubles

```python
name = 'idrees'
print(name)

name = "idrees"
print(name)

doc = "idrees is my name. teaching is my prof. and playing is my game."
print(doc)

# \n - newline character
doc = "idrees is my name.\n teaching is my prof. \n and playing is my game."
print(doc)

'e'
```

```python
# write multi line strings, use triple quotes and you will not even need the \n
doc = '''
idrees is my name. 
teaching is my prof. 
and playing is my game.
'''
print(doc)
```
