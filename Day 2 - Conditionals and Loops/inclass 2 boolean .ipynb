## can you compare or OPERATE multiple conditions, to get an output?

## BOOLEAN OPERATIONS

- not
- or
- and

**OR**

```python
# false   false
10 < 5 or 10 < 5  # False

# true   true
10 > 5 or 10 > 5  # True

# true   false
10 > 5 or 10 < 5  # True
```

**AND**

```python
# false   false
10 < 5 and 10 < 5  # False

# true   true
10 > 5 and 10 > 5  # True

# true   false
10 > 5 and 10 < 5  # False
```

**NOT**

```python
10 == 10 # True

10 == 9 # False

# not
10 != 9 # True
```
