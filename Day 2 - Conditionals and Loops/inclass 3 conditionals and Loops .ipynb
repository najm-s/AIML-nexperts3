# CONDITIONALS

- conditions for our programs to run

- you can choose whether you run a program or not based on, if a certain condition is being met

- Conditions are implemented using conditional statements:
    - if
    - elif
    - else

- syntax:

``` python
if (condition):
    <code goes here>
```


````python
print("I am Idrees")

if 5 == 6:
    print("I am from Kashmir!")


if 5 == 6:
    print("I am from Kashmir!")
elif 5 == 12/2:
    print("I am from Kuala Lumpur")
else:
    print("be kati aaas!!") # where did i come from

````

## IF ELSE Ladder

````python
# You want to create a small program that assigns grades to people with different marks
# Grade calculator 

marks = 45

if marks > 90:
    grade = 'A+'
elif marks > 80:
    grade = 'A'
elif marks > 70:
    grade = 'B'
elif marks > 55:
    grade = 'C'
elif marks >= 45:
    grade = 'D'
else:
    grade = 'F'

print("The grade is: ",grade)
````

# **LOOPs**

Loops, help you execute programs repeatedly
- help you repeat actions, without repeating the code
- Loops are iterations too



- Two main loop types in python:
    - for
    - while


```python
for <variable> in <sequence>:
    # code block 
```


- Variable you use, **iterator variable**
- after every code execution, its value is updated

````python
for x in range(5):
    print(x)
````

## WHILE
- Keep an exit condition, before you **execute**

````python
# create a counter

count = 10

while count > 0:
    print("Count: ",count)
    count -= 1 # count = count - 1
````

````python
# Rocket Launch Timer

import time # inbuilt module in python

count = 10

while count > 0:
    print("T minus: ",count)
    count -= 1 # count = count - 1
    time.sleep(1) # wait 1 second
````

Loops, we need ways of getting out of the loop

```python
while 1:
    print("this will run infinity times!!!")
```

- break - helps us do that 

````python
count =  1
while count < 10:
    print("T minus: ",count)
    count += 1 # count = count - 1
    if count == 6:
        break
        print("after break")
        print("after break")
        print("save secrets!!!")

print("out of while, after break")
````
