## String methods

````python
name = "   I AM idrees   "

print(name)

# string methods
# any variable of string type. You can do a lot with it
# using the . dot operator - helps us access the string operations, or string methods

print(name.upper())
````

````python
name.upper() # I AM IDREES

name.lower() # i am idrees

name.title() # I Am Idrees

name.strip() # removes the leading and trailing spaces

name.lstrip() # LEFT - leading spaces erased = spaces in the beginning 

name.rstrip() # RIGHT - Trailing spaces erased = spaces in the end
````

this was the string:
```
"    I  A M    i  d  r  e  e  s   "
0123 4  5 6 78 9 10 11 12 13 14
```
- every single character in the string has a number - INDEX

````python
# find things in a string
name.find("i") # found after 8 index - find gives the index of the first occurence

name.find("e")

name.replace("idrees","Adilah")

name.replace(" ","") # take empty spaces and remove them

name.replace(" ","-")

name.count("e") # how many substring there are in the string

name.split() # we get a list with words tokenised


# In LLMs like ChatGPT etc
# TOKENs - pay 10 dollars, 500K tokens 

# prompt - "create an image of a horse on a bicyle"
prompt = "create an image of a horse on a bicycle"
prompt.split()
prompt.count("bicycle")
````

### First intro to how NLP works
- or **NATURAL LANGUAGE PROCESSING**

- AI capability to understand and generate human or natural language

````python
message = '''Python is a versatile programming language. It is widely used in various domains such as web development, 
data science, artificial intelligence, and more. 
Its simplicity and readability make it a popular choice among developers of all levels. Whether you are a
beginner or an experienced programmer, Python offers a wealth of libraries and frameworks to help you achieve 
your goals.
We can do a lot with Python, from building web applications to analyzing data and creating machine learning models. 
Its extensive standard library and vibrant community contribute to its popularity and effectiveness in solving 
real-world problems.
'''
````

````python
# 1. Replace all occurrences of "Python" with "PYTHON" in the message string
message.replace("Python","PYTHON")
# 2. Find out if the message contains the word "data"
message.find("data")
# 3. Replace all occurrences of spaces with underscores (_)
message.replace(" ","_")
# 4. Find the location of the first occurrence of the word "programming"
message.find("programming")
# 5. Split the message into a list of sentences
message.split(".") # sentence tokenization
# we are using . as a delimiter
# something that we use to split data into parts
````

````python
# Data Cleaning

message = message.replace(" ","")
message = message.replace("\n","")
message = message.replace(".","")
message = message.replace(",","")
message = message.replace("-","")
message = message.lower()
print(message)
````
