# CONDITIONALS

- conditions for our programs to run

- you can choose whether you run a program or not based on, if a certain condition is being met

- Conditions are implemented using conditional statements:
    - if
    - elif
    - else

- syntax:

``` python
if (condition):
    <code goes here>
```


````python
print("I am Idrees")

if 5 == 6:
    print("I am from Kashmir!")


if 5 == 6:
    print("I am from Kashmir!")
elif 5 == 12/2:
    print("I am from Kuala Lumpur")
else:
    print("be kati aaas!!") # where did i come from

````

````python
x = 10
y = 10

if x == y:
    print("I am from Kashmir!")
elif x > y:
    print("I am from Kuala Lumpur")
else:
    print("be kati aaas!!") # where did i come from

# Output: 
````

## IF ELSE Ladder

````python
# You want to create a small program that assigns grades to people with different marks
# Grade calculator 

marks = 45

if marks > 90:
    grade = 'A+'
elif marks > 80:
    grade = 'A'
elif marks > 70:
    grade = 'B'
elif marks > 55:
    grade = 'C'
elif marks >= 45:
    grade = 'D'
else:
    grade = 'F'

print("The grade is: ",grade)
````

# **LOOPs**

Loops, help you execute programs repeatedly
- help you repeat actions, without repeating the code


- Two main loop types in python:
    - for
    - while


```python
for <variable> in <sequence>:
    # code block 
```

````python
for x in range(5):
    print(x)
````

````python
range(5) # creates a range object of 0 to 5
# 0 1 2 3 4

# iterable object

type(range(5))
````
